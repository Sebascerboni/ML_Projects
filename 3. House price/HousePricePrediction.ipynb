```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Importing the Boston House Price Dataset

```python
house_price_dataset = sklearn.datasets.load_boston()
```

```python
print(house_price_dataset)
```

```python
# Loading the dataset to a pandas dataframe
house_price_dataframe = pd.DataFrame(house_price_dataset.data)
print(house_price_dataframe)
```

```python
house_price_dataframe = pd.DataFrame(house_price_dataset.data, columns=house_price_dataset.feature_names)
```

```python
house_price_dataframe.head()
```

```python
# adding the target column
house_price_dataframe['price'] = house_price_dataset.target
```

```python
house_price_dataframe.head()
```

```python
house_price_dataframe.shape
```

```python
#Checking missing values
house_price_dataframe.isnull().sum()
```

```python
#Staticals measures
house_price_dataframe.describe()
```

Understanding the correlation between the features in the dataset
1. **Positive Correlation:** if one variable affects directly to other (increase)
2. **Negative Correlation:** if one value decrease other

```python
correlation = house_price_dataframe.corr()
```

```python
# Making a heat map (Graphic used to understand the correlations between varaibles)
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.2f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

Splitting data into: training and testing. And data and target 

```python
X = house_price_dataframe.drop('price', axis=1)
Y = house_price_dataframe['price']
```

```python
print(X)
```

```python
print(Y)
```

Testing and Training data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

Model Training

XGBoost Regressor

```python
#loading the model
model = XGBRegressor()
```

```python
#Training the model
model.fit(X_train, Y_train)
```

Evaluation

Prediction on training data

```python
# accuracy for prediction on training data
training_data_prediction = model.predict(X_train)
```

```python
#R squared error
score_1 = metrics.r2_score(Y_train, training_data_prediction)
#Mean absolute error
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)
print("R-squared error: ",score_1)
print("Mean Absolute error: ",score_2)
```

Visualating the actual prices and predicted prices

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel('Actual prices')
plt.ylabel('Predicted prices')
plt.title('Actual Price VS Predicted Price')
plt.show()
```

Prediction on testing data

```python
# accuracy on testing data prediction
testing_data_prediction = model.predict(X_test)
#R squared error
score_1 = metrics.r2_score(Y_test, testing_data_prediction)
#Mean absolute error
score_2 = metrics.mean_absolute_error(Y_test, testing_data_prediction)
print("R-squared error for testing: ",score_1)
print("Mean Absolute error for testing: ",score_2)
```

```python
plt.scatter(Y_test, testing_data_prediction)
plt.xlabel('Actual prices')
plt.ylabel('Predicted prices')
plt.title('Actual Price VS Predicted Price')
plt.show()
```
